```python
import pandas as pd
import matplotlib.pyplot as plt

# Carrega as bases
df_treino = pd.read_parquet("../data/processed/base_train.parquet")
df_producao = pd.read_parquet("../data/01_raw/dataset_kobe_prod.parquet")


colunas = ["lat", "lon", "minutes_remaining", "period", "playoffs", "shot_distance"]


fig, axes = plt.subplots(nrows=3, ncols=2, figsize=(14, 10))
axes = axes.flatten()  

for i, col in enumerate(colunas):
    axes[i].hist(df_treino[col], bins=20, alpha=0.5, label='Treino')
    axes[i].hist(df_producao[col], bins=20, alpha=0.5, label='Produo')
    axes[i].set_title(f'Distribuio da feature: {col}')
    axes[i].legend()

plt.tight_layout()
plt.show()
```
